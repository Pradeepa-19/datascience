```python
#bar chart
p=['food','cloths','electronic','home appl']
q=[1500,7300,4500,5041]
plt.bar(p,q,color='brown')
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
shares=[35,45,55,30]
product=['redmi','samsung','apple','vivo']
plt.pie(shares,labels=product,colors=['brown','red','pink','blue'],autopct='%1.1f%%')
plt.show()
```

```python
#histogram
marks=[35,45,50,70,95,99]
plt.hist(marks,color='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
t=pd.read_csv("titanic.csv")
plt.hist(t['Age'])
plt.show()
```

```python
plt.bar(t['Sex'],t['Survived'])
plt.xlabel('Sex')
plt.ylabel('Survived')
plt.title('Gender vs survived')
plt.show()
```

```python
#scatter plot
plt.scatter(t['Age'],t['Fare'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
t=pd.read_csv("titanic.csv")
t['Age'].fillna(t['Age'].mean())
```

```python
t.boxplot(column='Age',by='Sex')
```

```python
#area plot
plt.fill_between(t['Age'],t['Fare'])
plt.show()
```

```python
#seaborn
import seaborn as sns
nd=t.select_dtypes(include=['int64','float64'])
sns.heatmap(nd.corr(),annot=True)
```

```python
#subplot

plt.subplot(2,2,1)
plt.scatter(t['Age'],t['Fare'])
plt.subplot(2,2,2)
plt.bar(t['Sex'],t['Survived'])
plt.figure(figsize=(4,3))
plt.show()
```
