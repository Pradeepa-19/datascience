```python
import numpy as np
marks=np.random.randint(0,101,500)
print(marks)
top=np.max(marks)
avg=np.mean(marks)
failed=marks[marks<35]
print(top)
print(avg)
print("failed",failed)
topper=marks[marks>90]
average=marks[(marks>=45) & (marks<=69)]
print("topper",topper)
print("avg",average)
```
