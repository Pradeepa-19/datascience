```python
import pandas as pd
d=pd.read_csv("titanic.csv")
d.drop(['Cabin','Embarked'],axis="columns")
```

```python
d['Age']=d['Age'].fillna(d['Age'].median())
d
```

```python
d.info()
```

```python
d=d.drop(['Name','Ticket'],axis='columns')
d
```

```python
d.groupby("Sex")["Survived"].mean()
```

```python
sd=pd.get_dummies(d['Sex'])
sd
```

```python
df=pd.concat([d,sd],axis='columns')
df
```

```python
df=df.drop(['PassengerId','Pclass','Sex','SibSp','Parch','Fare'],axis='columns')
df
```
